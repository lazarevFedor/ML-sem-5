### ***3. Анализ атрибутов.***
***3.1 Определение среднего значения и СКО***

```python
cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in cols:
    mean_value = df[col].mean()
    std_value = df[col].std()
    print(f'Среднее значение для {col}: {mean_value:.2f}')
    print(f'СКО для {col}: {std_value:.2f}\n')
```

***3.2 Построить гистограмму распределения значений. Определить есть ли вбросы.***

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')

list_of_outliers = []

for col in cols:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], bins=15)
    plt.title(f'Гистограмма распределения значений {col}:')
    plt.xlabel(col)
    plt.ylabel('count')
    plt.show()

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)]
    num_outliers = len(outliers)
    list_of_outliers.append(num_outliers)

sns.countplot(x=df['type'])
plt.title('Гистограмма распределения значений type')
plt.show()

print('\n\nСводка по выбросам в данных:\n')
i = 0
for col in cols:
    print(f'Количество выбросов для {col} = {list_of_outliers[i]}')
    i += 1
```

***3.3 Определение пропущенных значений, их количество и обработка.***

```python
df_clean = df.copy()

for col in cols:
    missing = df[col].isnull().sum()
    missing_pct = (missing / len(df)) * 100
    print(f'Количество пропущенных значений для {col} = {missing} ({missing_pct:.2f}%)')
    if missing_pct >= 5:
        df_clean[col].fillna(df_clean[col].median(), inplace=True)
    else:
        df_clean.dropna(subset=[col], inplace=True)

missing = df['type'].isnull().sum()
missing_pct = (missing / len(df)) * 100
print(f'Количество пропущенных значений для type = {missing} ({missing_pct:.2f}%)')
print('\nКоличество пропущенных значений после обработки:')
print(df_clean.isnull().sum())
```

Обработку пропущенных значений произвели следующим образом:

Если пропусков мало(меньше 5 %), то будем их удалять, в противном случае будем заполнять пропуски, чтобы сохранить объем данных, в численных данных будем заполнять медианой, так как в них имеются выбросы, к которым она устойчива.

### ***4. Определение корреляции между параметрами.***
***4.1 Определение высококоррелированных атрибутов и характера корреляции***

Корреляцию будем измерять для численных признаков, поэтому исключаем атрибут type из рассмотрения в этом пункте, так как это категориальный признак.

```python
num_df = df_clean.select_dtypes(include=['float64', 'int64'])
corr_matrix = num_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Матрица корреляций')
plt.show()

print('Высококоррелированные атрибуты:')
high_corr = [(i, j, corr_matrix.loc[i,j]) for i in corr_matrix.columns for j in corr_matrix.columns if i < j and abs(corr_matrix.loc[i,j]) > 0.6]
for i in high_corr:
    print(f'- Корреляция между {i[0]} и {i[1]} = {i[2]:.2f}')
```

Характер корреляции:

- чем больше free sulfur dioxide, тем больше total sulfur dioxide. Корреляция положительна.
- чем больше free alcohol, тем меньше density. Корреляция положительна.

***4.2 Определение атрибутов, не имеющих корреляцию***

```python
print('\nНе коррелирующие атрибуты:')
for i in corr_matrix.columns:
    for j in corr_matrix.columns:
        if i < j and abs(corr_matrix.loc[i,j]) < 0.01:
            print(f'- {i} и {j}')
```

***4.3 Построение графиков рассеивания***

```python
for i in high_corr:
    sns.scatterplot(x=num_df[i[0]], y=num_df[i[1]]); plt.title(f'Рассеивание {i[0]} и {i[1]}')
    plt.show()
```

***4.4 Анализ полученных результатов***

Проведем анализ полученных результатов. После проделанной работы можно выявить следующее:

- Больше всего на качество вина (атрибут quality) в положительную сторону повлияло количество алкоголя в нем(атрибут alcohol). Корреляция между ними составила 0.44. В негативную же сторону повлияла летучая кислотность (атрибут volatile acidity) c корреляцией -0.27. Это объяснимо тем, что у кислого вина будет менее приятный аромат, а высокое содержание алкоголя дает более интенсивный вкус.
- Высокую корреляцию имели атрибуты free sulfur dioxide и total sulfur dioxide, что вполне логично, число свободных диоксидов серы влияет на общее число диоксидов серы в вине. Также скоррелированы были атрибуты алкоголь (атрибут alcohol) и плотность(атрибут density). Это объяснимо так как вода плотнее, чем этанол, следовательно, при высоком содержании алкоголя, плотность будет меньше.
- Наименее скоррелированы атрибуты alcohol и sulphates. Это действительно так, так как сульфаты добавляют на разных этапах производства, не связанных с алкоголем, в качестве антиоксиданта. Алкоголь пускай и оказывает похожее антимикробное действие на вино, но в сильно меньшей степени, чем сульфаты.
