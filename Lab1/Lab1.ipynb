### ***1. Выбор исследуемого датасета.***
В качестве исследуемого был выбран датасета с сайта [_kaggle_](https://www.kaggle.com/datasets/rajyellow46/wine-quality "Тыкни на меня").

### ***2. Краткое описание датасета.***
***2.1 Предметная область, источник данных, характер данных***
- Предметная область: Рассматриваемые данные представляют собой физико-химические свойства, тип и оцененное качество Португальского вина "Vinho Verde".
- Источник данных: [_kaggle_](https://www.kaggle.com/datasets/rajyellow46/wine-quality "Тыкни на меня").
- Характер данных: Данные реальные, получены из физико-химических измерений и сенсорных оценок реальных образцов вина.

***2.2 Атрибуты представленные в датасете и их тип***

В датасете представлены следующие атрибуты:
1. fixed acidity - фиксированная кислотность - числовой тип
2. volatile acidity - летучая кислотность - числовой тип
3. citric acid - количество лимонной кислоты - числовой тип
4. residual sugar - остаточный сахар - числовой тип
5. chlorides - хлориды - числовой тип
6. free sulfur dioxide - свободный диоксид серы - числовой тип
7. total sulfur dioxide - общий диоксид серы - числовой тип
8. density - плотность - числовой тип
9. pH - мера кислотности - числовой тип
10. sulphates - количество сульфитов - числовой тип
11. alcohol - процент алкоголя - числовой тип
12. type - тип вина - строковый тип
13. quality - качество вина - числовой тип

***2.3 Описание задачи***

Датасет подходит для задачи регрессии -- предсказание качества вина(атрибут quality) на основании атрибутов 1-11. Также датасет подходит для задачи классификации -- определение типа вина(атрибут type) c метками классов red для красного вина и white для белого вина на основании атрибутов 1-11. Данные упорядочены, несбалансированны(средних вин больше, чем отличных или плохих). Возможен анализ на выбросы для редких оценок.
